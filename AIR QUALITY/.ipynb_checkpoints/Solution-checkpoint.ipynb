# Import Libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

# Load Air DataFrame

```python
df=pd.read_excel("AirQualityUCI.xlsx",na_values="-200")
df.head()
```

# Data Cleaning(Handling Missing Values)

```python
df.isnull().sum()
```

```python
for column in df.drop(["Date","Time"],axis=1).columns:
    df[column]=df[column].fillna(df[column].median())
```

## MAKING CO MY TARGET

```python
# Convert the 'time' column to datetime (with only time information)
df['Time'] = pd.to_datetime(df['Time'], format='%H:%M:%S')

# Extract the hour using .dt accessor
df['Hour'] = df['Time'].dt.hour

df["Date"]=pd.to_datetime(df["Date"])

df["Day"]=df["Date"].dt.day

df["Month"]=df["Date"].dt.month
```

```python
df=df.drop(["Date","Time"],axis=1)
```

```python
df.corr()
```

```python
X=df.drop(["T","RH","AH","Day","Month","CO(GT)"],axis=1)
y=df["CO(GT)"]
```

```python
df.describe()
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=42,test_size=0.3)
```

```python
columns_to_scale=X.columns[X.columns !="Hour"]
columns_to_scale
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler

pipe=Pipeline([
    ("scaler",StandardScaler()),
    ("poly",PolynomialFeatures(degree=2)),
    ("model",LinearRegression())
])

pipe.fit(X_train,y_train)
pipe.score(X_test,y_test)
```

```python
from sklearn.ensemble import BaggingRegressor
from sklearn.tree import DecisionTreeRegressor
pipe=Pipeline([
    ("scaler",StandardScaler()),
    ("poly",PolynomialFeatures(degree=2)),
    ("model",BaggingRegressor(estimator=DecisionTreeRegressor()))
])

pipe.fit(X_train,y_train)
pipe.score(X_test,y_test)
```

```python
from xgboost import XGBRegressor

pipe=Pipeline([
    ("scaler",StandardScaler()),
    ("poly",PolynomialFeatures(degree=2)),
    ("model",XGBRegressor())
])
pipe.fit(X_train,y_train)
pipe.score(X_test,y_test)
```
