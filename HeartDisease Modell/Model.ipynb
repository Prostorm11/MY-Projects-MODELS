```python
from ucimlrepo import fetch_ucirepo 
import pandas as pd
  
# fetch dataset 
heart_disease = fetch_ucirepo(id=45) 
  

```

```python
heart_disease.data.features
```

```python
heart_disease.data.targets
```

```python
df=heart_disease.data.original
df
```

```python
df.describe()
```

```python
df["ca"]=df["ca"].fillna(df["ca"].median())
df["thal"]=df["thal"].fillna(df["thal"].median())
```

```python
X=df.drop("num",axis=1)
y=df["num"]
```

## TRAINING WITH THE LOGISTIC MODEL

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=42,test_size=0.25)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report

model=LogisticRegression(max_iter=1000,C=0.1)
model.fit(X_train,y_train)
y_pred=model.predict(X_test)
report=classification_report(y_test,y_pred)
print(report)
```

## TRAINING WITH DECISIION TREES

```python
from sklearn.tree import DecisionTreeClassifier

model=DecisionTreeClassifier()
model.fit(X_train,y_train)
y_pred=model.predict(X_test)
report=classification_report(y_test,y_pred)
print(report)
```

## TRAINING WITH LOGISTIC REGRESSION AND ALSO PIPELINE SCALING

```python
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline

pipe_model=Pipeline([
    ("scaler",StandardScaler()),
    ("model",LogisticRegression(C=0.1))
])

pipe_model.fit(X_train,y_train)
```

```python
y_pred=pipe_model.predict(X_test)
report=classification_report(y_test,y_pred)
print(report)
```

```python
from xgboost import XGBClassifier
model=XGBClassifier()
model.fit(X_train,y_train)
y_pred=model.predict(X_test)
report=classification_report(y_test,y_pred)
print(report)
```

```python
from imblearn.under_sampling import RandomUnderSampler
from imblearn.over_sampling import SMOTE
from imblearn.combine import SMOTETomek

rus=RandomUnderSampler()
smotek=SMOTETomek()
X_rus,y_rus=rus.fit_resample(X_train,y_train)
X_train_smotek,y_train_smotek=smotek.fit_resample(X_train,y_train)
y_rus.value_counts()
```

```python
model=LogisticRegression(C=0.01,max_iter=1000)
model.fit(X_rus,y_rus)

y_pred=model.predict(X_test)
score=classification_report(y_test,y_pred)
print(score)
```

```python
model=LogisticRegression(C=0.01,max_iter=1000)
model.fit(X_train_smotek,y_train_smotek)

y_pred=model.predict(X_test)
score=classification_report(y_test,y_pred)
print(score)
```

```python
model=DecisionTreeClassifier(criterion="gini",)
model.fit(X_train_smotek,y_train_smotek)
y_pred=model.predict(X_test)
score=classification_report(y_test,y_pred)
print(score)
```

```python
from sklearn.svm import SVC

model=SVC(kernel="rbf",C=4.00001)
model.fit(X_train_smotek,y_train_smotek)
y_pred=model.predict(X_test)
score=classification_report(y_test,y_pred)
print(score)
```
