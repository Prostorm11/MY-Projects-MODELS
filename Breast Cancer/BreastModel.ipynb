```python
import pandas as pd
from ucimlrepo import fetch_ucirepo 
  
# fetch dataset 
breast_cancer_wisconsin_diagnostic = fetch_ucirepo(id=17) 
breast_cancer_wisconsin_diagnostic
```

```python
df=breast_cancer_wisconsin_diagnostic.data.original
```

```python
df
```

```python
df.isnull().sum()
```

```python
import seaborn as sns

sns.barplot(df.Diagnosis.value_counts())
```

```python
df=pd.get_dummies(df,columns=["Diagnosis"],drop_first=True)
```

```python
df
```

```python
df.corr()
```

```python
from sklearn.model_selection import train_test_split

X=df.drop(["ID","Diagnosis_M"],axis=1)
y=df.Diagnosis_M

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report

pipe=Pipeline([
    ("scaler",StandardScaler()),
    ("model",LogisticRegression())
])
pipe.fit(X_train,y_train)
y_pred=pipe.predict(X_test)
report=classification_report(y_test,y_pred)
print(report)
```
