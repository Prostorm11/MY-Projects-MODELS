```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("MathE dataset (4).csv", delimiter=";", encoding="ISO-8859-1")
df
```

## Data Evaluation Preperation And Feature Engineering

```python
df.isna().sum()
```

```python
df=df.drop("Student ID",axis=1)
df
```

```python
df["Type of Answer"].value_counts()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vector=CountVectorizer()
X=df.drop(["Type of Answer","Keywords","Question ID"],axis=1)
y=df["Type of Answer"]
v=vector.fit_transform(df.Keywords)
df_v=pd.DataFrame(v.toarray(),columns=vector.get_feature_names_out())
X=pd.get_dummies(X,drop_first=True)
```

```python
X=pd.concat([X,df_v],axis=1)
X
```

## Model Selection And Training

```python
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import GridSearchCV
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
```

```python
fold=StratifiedKFold(n_splits=6,shuffle=True,random_state=42)
models=[MultinomialNB(),RandomForestClassifier(),LogisticRegression()]
for i in models:
    cvs=cross_val_score(i,X,y,cv=fold)
    print({"model":i,"score":cvs})
```
